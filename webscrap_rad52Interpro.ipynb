```python
#Web scrapping for the entry 'Rad52 family' at the Alphafold page in Interpro databank
from urllib import request
import json
from time import sleep
```

```python
url = 'https://www.ebi.ac.uk/interpro/wwwapi//protein/UniProt/entry/InterPro/IPR041247/?page_size=100&has_model=true'
```

```python
rad52 = []

while url is not None:
    req = request.Request(url)
    response = request.urlopen(req)
    enconded_response = response.read()
    decoded_response = enconded_response.decode()
    payload = json.loads(decoded_response)

    for item in payload['results']:
       print(item['metadata']['accession'], item['metadata']['name'], item['metadata']['length'], item['metadata']['source_organism']['scientificName'])
       #Append the results to the list
       rad52.append([item['metadata']['accession'], item['metadata']['name'], item['metadata']['length'], item['metadata']['source_organism']['scientificName']])
    url = payload['next']
    sleep(1)
    
```

```python
len(rad52)
```

```python
#Creating a table from sak3 data
import pandas as pd
pd.DataFrame()
pd.DataFrame()

table_rad52 = pd.DataFrame(rad52)
table_rad52.head()
```

```python
#Adjust column names
table_rad52.columns = ['Accession', 'Protein Name', 'Length', 'Organism Scientific Neme']
table_rad52.head()
```

```python
#Saving the table as a .csv file
table_rad52.to_csv('rad52.csv', index=False)
```
